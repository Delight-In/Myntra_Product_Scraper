```python
# Import required package/library
import os
import time
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
Product = 'kids footwear'
driver = webdriver.Chrome()
driver.get("https://www.myntra.com/dress?rawQuery="+Product)

text = driver.page_source
html_text = bs(text, "html.parser")
```

```python
text
```

```python
html_text
```

```python
search = html_text.find_all("ul",{"class":"results-base"})
search[0]
```

```python
for i in search:
    extract = i.find_all('a', href = True)
    p = (extract[0]['href'])

product_url = "https://www.myntra.com/"+p
```

```python
product_url
```

```python
driver.get(product_url)
s = driver.page_source
s
```

```python
s_html = bs(s, 'html.parser')
s_html
```

```python
tital_fetch = s_html.findAll('title')
tital_fetch
```

```python
title = tital_fetch[0].text
title
```

```python
overallRating = s_html.findAll('div',{'class':'index-overallRating'})
overallRating
```

```python
for i in overallRating:
    rating = i.find('div').text
rating
```

```python
price = s_html.findAll('span',{'class':'pdp-price'})
price
```

```python
for i in price:
    price = i.find('strong').text

price
```

```python
detail_review = s_html.find('a', {'class':'detailed-reviews-allReviews'})
detail_review
```

```python
review = "https://www.myntra.com" + detail_review['href']
review
```

```python
driver.get(review)
review_page = driver.page_source
```

```python
review_page
```

```python
driver.quit()
```

```python
review = bs(review_page,"html.parser")
review
```

```python
reviews = review.find('div',{'class':'detailed-reviews-userReviewsContainer'})
reviews
```

```python
for i in reviews:
    user_rating = i.findAll('div',{'class':'user-review-main user-review-showRating'})
    user_comment = i.findAll('div',{'class':'user-review-reviewTextWrapper'})
    user_name = i.findAll('div',{'class':'user-review-left'})

print(user_rating)
print(user_comment)
print(user_name)
```

```python
for i in user_rating:
    rate = i.find('span',{'class':'user-review-starRating'}).text

rate
```

```python
for i in user_comment:
    comment = i.text

comment
```

```python
for i in user_name:
    name = i.find('span').text

name
```

```python
for user in user_name:
    # Find all spans within the current user_name div
    spans = user.find_all('span')
    if len(spans) > 1:  # Check if there are at least two spans
        second_span_text = spans[1].text  # Get the text of the second span
        print(second_span_text)  # Output: Second Span
```
